```python
#Take an input from the user for start and end values. Print all even numbers between them.
start = int(input("start:"))
end = int(input("stop:"))
loop_count =0
while start <=end:
  loop_count =loop_count+1
  if start%2 ==0:
    print(start)
  start =start+1
print("loop count", loop_count)
```
