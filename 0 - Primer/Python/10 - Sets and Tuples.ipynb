```python
def check_binary(a):
  for i in a:
    if not (i == "0" or i =="1"):
      return "not binary"
  return "binary"
```

```python
check_binary("0100101010109")
```

```python
def check_binary(a):
  s = set(a)
  if s == {"0", "1"}:
    return "binary"
  else:
    return "not binary"
```

```python
check_binary("0101")
```
