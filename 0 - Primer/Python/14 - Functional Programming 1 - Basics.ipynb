```python
#dictionary of students

students = [
    {"name": "A", "marks": 50},
    {"name": "B", "marks": 100},
    {"name": "C", "marks": 40},
    {"name": "D", "marks": 70},
    {"name": "E", "marks": 60},
]
```

```python
students[0] > students[1]
```

```python
get_value= lambda x: x["marks"]
```

```python
get_value(students[0]) >get_value(students[1])
```

```python
sorted(students)
```

```python
sorted(students, key= lambda x: x["marks"])
```

```python
sorted(students, key= lambda x: x["marks"], reverse = True)
```

```python
def pretty(func):
  def inner():
    print("*" * 50)
    func()
    print("*" * 50)

  return inner
```

```python
stars= pretty(boo)
```

```python
stars()
```
