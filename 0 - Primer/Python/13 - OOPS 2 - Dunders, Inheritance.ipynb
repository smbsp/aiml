```python
class Car:
    def __init__(self, name, mileage):
        self.name = name
        self.mileage = mileage
    # Whenever an object is printed, __str__() function is called.
    def __str__(self):
      return f"Suraaj is great but the mileage is {self.mileage} for {self.name}"

    def __add__(self, other):
      return self.mileage + other.mileage

    def __lt__(self, other):
      return self.mileage < other.mileage
```

```python
c1 = Car("Nexon", 12)
c2= Car("Venue", 15)
```

```python
c1+c2
```

```python
c1<c2
```

```python
#inheritance


class SchoolMember:
  def __init__(self, name):
    self.name = name

class Student (SchoolMember):
  def __init__(self, name, grade):
    super().__init__(name)
    self.grade = grade

class Staff (SchoolMember):
  def __init__(self, name, salary):
    super().__init__(name)
    self.salary = salary

class Teacher (Staff):
  def __init__(self, name,salary , subject):
    super().__init__(name, salary)
    self.subject = subject

```

```python
s1= Student("Suraaj", "B")
```

```python
type(s1)
```

```python
s1.grade
```

```python
s1.name
```

```python
t1= Teacher("ABC", 20000, "English")
```

```python
t1.name
```

```python
t1.salary
```

```python
t1.subject
```

```python
class A:
    def __init__(self, a):
        self.a = a

class B:
    def __init__(self, b):
        self.b = b

class C (A,B):
  def __init__(self, a,b,c):
    A.__init__(self, a)
    B.__init__(self, b)
    self.c= c


```

```python
c1= C(1,2,3)
```

```python
c1.a
```

```python
c1.b
```

```python
c1.c
```
