```python
# 1) Define a class (this creates a *class object* at runtime)
class Car:
    # 2) Constructor that stores instance attributes in the instance __dict__
    def __init__(self, name, age):
        self.name = name  # 3) Stored in instance attribute dict (usually)
        self.age = age    # 4) Stored in instance attribute dict (usually)

    # 5) A normal instance method
    def drive(self):
        print("VROOOOMMMMM!")
```

```python
# 6) Create an instance (this creates a *Car instance object* at runtime)
c = Car("Seltos", 3)
```

```python

# 7) Helper: show an object's id in decimal and hex (hex helps match 0x... style)
def show(label, obj):
    # 8) id(obj) is unique during the object's lifetime (CPython: it's the address)
    print(f"{label:22} id(dec)={id(obj)}  id(hex)={hex(id(obj))}  type={type(obj)}")
```

```python
print("\n--- Core objects ---")
show("Car (the class)", Car)          # 9) The class itself is an object
show("type(Car)", type(Car))          # 10) Usually <class 'type'> (metaclass)
show("c (instance)", c)               # 11) The instance is an object
show("type(c)", type(c))              # 12) Should be <class '__main__.Car'> (i.e., Car)

print("\n--- Attribute storage ---")
show("c.__dict__", c.__dict__)        # 13) Dict holding instance attributes (if present)
print("c.__dict__ =", c.__dict__)     # 14) You should see {'name': 'Seltos', 'age': 3}

show("c.name", c.name)                # 15) The string object "Seltos"
show("c.age", c.age)                  # 16) The int object 3 (often cached for small ints)

print("\n--- Class namespace (methods live here) ---")
show("Car.__dict__", Car.__dict__)    # 17) MappingProxy over class attributes (methods, etc.)
print("Is 'drive' stored on class?", "drive" in Car.__dict__)  # 18) True
show("Car.drive (function)", Car.drive)  # 19) Function object stored on the class

print("\n--- Bound vs unbound method ---")
show("c.drive (bound)", c.drive)      # 20) Bound method object (wraps function + instance)
# 21) Calling it shows that `c` is implicitly passed as the first arg (self)
c.drive()

print("\n--- Where does c point for its class? ---")
show("c.__class__", c.__class__)      # 22) Same object as Car
print("c.__class__ is Car ?", c.__class__ is Car)  # 23) Should be True
```

```python
print(type(Car) is type)
print(type(c) is Car)
print(c.drive.__self__ is c)
print(c.drive.__func__ is Car.drive)
print(id(c.__dict__), id(c.name), id(c.age))
```
