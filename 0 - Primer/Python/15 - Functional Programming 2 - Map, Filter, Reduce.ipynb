```python
#
heights = [144, 167, 189, 170, 190, 150, 165, 178, 200, 130]

#h < 150 --> "S"
#h >= 150 and h < 180 --> "M"
#h >= 180 --> "L"
```

```python
def complex_logic(height):
    if height < 150:
        return "S"
    elif height >= 150 and height < 180:
        return "M"
    else:
        return "L"
```

```python
list(map(complex_logic, heights))
```

```python
list(map(lambda x: "S" if x<150 else "M" if x>=150 and x<180 else "L", heights))
```

```python
def sum_number(x, y, *args):
  print("args has: ", args)
  result=  x + y
  if args:
    result = result + sum(args)

  return result
```

```python
def create_person(name, age, gender):
    Person = {
        "name": name,
        "age": age,
        "gender": gender
    }

    return Person
```

```python
create_person("Suraaj", "99", gender="male")
```

```python
create_person("Suraaj", "99", gender="male", hobby="chess")
```

```python
def create_person(name, age, gender, **extra_info):
    Person = {
        "name": name,
        "age": age,
        "gender": gender
    }
    print("kwargs has: ",extra_info)
    if extra_info:
      Person.update(extra_info)
    return Person
```

```python
create_person("Suraaj", "99", gender="male", hobby="chess")
```

```python
create_person("Suraaj", "99", gender="male", hobby="chess", profession="Computers")
```
